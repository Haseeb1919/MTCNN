```python
from mtcnn import MTCNN
import cv2
```

```python
detector = MTCNN()
```

```python
#
cap=cv2.VideoCapture(0)

if(cap.isOpened==False):
    print ("the camera is not opened")

while True:

    ret, frame=cap.read()

    #deecting faces in every frame
    output= detector.detect_faces(frame)
    for face in output:
        x,y,w,h=face['box']
        #creating a rectangle around the face
        cv2.rectangle(frame,(x,y),(x+w,y+h),color=(0,255,0),thickness=2)
        #detecting the keypoints of face coordinates
        xleft_eye,yleft_eye=output[0]['keypoints']['left_eye']
        xright_eye,yright_eye=output[0]['keypoints']['right_eye']

        xnose,ynose=output[0]['keypoints']['nose']
        
        xmouth_left,ymouth_left=output[0]['keypoints']['mouth_left']
        xmouth_right,ymouth_right=output[0]['keypoints']['mouth_right']




    #creating circle around and keypoints of face
        cv2.circle(frame,center=(xleft_eye,yleft_eye),radius=2,color=(0,255,0),thickness=2)
        cv2.circle(frame,center=(xright_eye,yright_eye),radius=2,color=(0,255,0),thickness=2)
        cv2.circle(frame,center=(xnose,ynose),radius=2,color=(0,255,0),thickness=2)
        cv2.circle(frame,center=( xmouth_left,ymouth_left),radius=2,color=(0,255,0),thickness=2)
        cv2.circle(frame, center=(xmouth_right,ymouth_right),radius=2,color=(0,255,0),thickness=2)




        cv2.putText(frame,'face',org=(x,y),fontFace=cv2.FONT_HERSHEY_SIMPLEX,fontScale=1,color=(0,255,0),thickness=2)

    cv2.imshow('frame',frame)
    if cv2.waitKey(1) & 0xFF == ord('x'):
        break
cv2.destroyAllWindows()
```
